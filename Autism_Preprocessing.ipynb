##Add advice

```python
# Rename the column
imputed_data.rename(columns={'A10_Autism_Spectrum_Quotient': 'A10'}, inplace=True)
```

```python
imputed_data.head()
```

```python
columns_Advice = [
    "A1_Advice", "A2_Advice", "A3_Advice",
    "A4_Advice", "A5_Advice", "A6_Advice", "A7_Advice",
    "A8_Advice", "A9_Advice", "A10_Advice"
]
```

```python
# Define a function to generate advice based on scores
def generate_advice(score, question):
    advice_dict = {
    "A1_Advice": "Encourage response to name through frequent and engaging verbal interactions.",
    "A2_Advice": "Practice eye contact exercises by making faces and playing eye-contact games.",
    "A3_Advice": "Use visual aids and interactive toys to motivate pointing gestures.",
    "A4_Advice": "Engage in activities that involve shared attention, such as reading books together.",
    "A5_Advice": "Incorporate pretend play scenarios into daily routines to foster imagination.",
    "A6_Advice": "Use your own gaze to guide the child’s attention to objects or events.",
    "A7_Advice": "Model and encourage comforting behaviors through role-playing and emotional support activities.",
    "A8_Advice": "Support language development with regular conversation and repetition of simple words.",
    "A9_Advice": "Introduce and reinforce the use of basic gestures, such as waving or pointing, during daily activities.",
    "A10_Advice": "Consult with a specialist if you notice frequent, purposeless staring to rule out developmental concerns."
}
    return advice_dict.get(question, 0) if score == 1 else "0"
```

```python
# Add advice columns to the DataFrame
for col in columns_Advice:
    # Map advice based on scores in the corresponding question column
    question_col = col.replace('_Advice', '')  # Remove '_Advice' to get the question column name
    imputed_data[col] = imputed_data.apply(lambda row: generate_advice(row[question_col], col), axis=1)
```

```python
imputed_data.head()
```

```python
# Create a new column 'Advice' that combines all non-zero advice
imputed_data['Advice'] = imputed_data[columns_Advice].apply(lambda row: ', '.join([advice for advice in row if advice != "0"]), axis=1)
```
